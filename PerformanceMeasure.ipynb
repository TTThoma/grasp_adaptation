```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import tensorflow as tf
```

```python
def plot_results(data_list,color='r',label='SERN'):
    mean_list = np.mean(data_list,axis=0)
    std_list = 0.5*np.std(data_list,axis=0)
    plt.fill_between(np.arange(1,len(mean_list)+1),mean_list-std_list,mean_list+std_list,color=color,alpha=0.3)
    plt.plot(np.arange(1,len(mean_list)+1),mean_list,color+'-',label=label)
```

```python
exp3_data=np.load("./EXP3-GQ-Marjoram_800_tex-results.npz")["arr_0"]
greedy_data=np.load("./Greedy-GQ-Marjoram_800_tex-results.npz")["arr_0"]

plot_results(greedy_data,color='r',label='Greedy')
plot_results(exp3_data,color='b',label='SERN')
plt.xlim([1,5])
plt.legend(fontsize=17)
plt.xlabel("Iterations",fontsize=17)
plt.ylabel("Success Rate",fontsize=17)
plt.show()
```
